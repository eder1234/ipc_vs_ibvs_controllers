```python
import numpy as np

def compute_trajectory_length(data):
    if len(data) < 2:
        return 0.0
    positions = [agent_state.position for agent_state in data]
    total_length = 0.0
    for i in range(len(positions) - 1):
        current_pos = positions[i]
        next_pos = positions[i + 1]
        delta = next_pos - current_pos
        distance = np.linalg.norm(delta)
        total_length += distance
    return total_length
```

```python
traj_len = []
file_len = []
for file in files:
    with open(file, 'rb') as f:
        data = pickle.load(f)
    file_len.append(len(data))
    traj = compute_trajectory_length(data)
    traj_len.append(traj)
```

```python
results_df['traj_len'] = traj_len
results_df['num_steps'] = file_len
results_df
```
